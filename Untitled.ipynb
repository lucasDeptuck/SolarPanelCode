```python
import numpy as np
import math
import time
import matplotlib.pyplot as plt
```

```python
def sundeclination(N):
    declination = -arcsin(0.39779 * cos(0.98565 * (N + 10) + 1.914 * sin(0.98565 * (N - 2))))
    return declination
```

```python
def greenwichmeantime():
    date_time = time.localtime()
    year = date_time[0]
    month = date_time[1]
    day = date_time[2]
    hour = date_time[3]
    
    return year, month, day, hour
```

```python
def rangecheck(value):
    print("starting value", value)
    num = 0
    while num == 0:
        if value < 0:
            value += 360
        if value > 360:
            value -= 360
        if value < 360 and value > 0:
            num = 1
        print("ending value", value) 
    return value    
```
